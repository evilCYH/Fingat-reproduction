```python
import numpy as np
import pandas as pd

df = pd.read_csv("./datasets4/arrange_sp500_companies.csv", encoding='ISO-8859-1')
company_list = df["Symbol"].unique()
sector_list = df["Sector"].unique()
```

# 构建行业间节点关系

```python
sector_list
```

```python
len(sector_list)
```

```python
outer_edge = []
for i in range(len(sector_list)):
    for j in range(i, len(sector_list)):
        outer_edge.append((i, j))
```

```python
company2index = {}
for i, key in enumerate(sector_list):
    company2index[key] = i
```

```python
company2index
```

```python
outer_edge
```

```python
len(outer_edge)
```

```python
outer_edge = np.array(outer_edge)
print(outer_edge.shape)
```

```python
np.save("./datasets4/outer_edge.npy", outer_edge)
```

# 行业内股票关系

```python
company_list
```

```python
len(company_list)
```

```python
name2index = {}
for i, key in enumerate(company_list):
    name2index[key] = i
```

```python
name2index
```

```python
inner_edge = []
for sector in sector_list:
    intra_companies = df[df.Sector == sector]["Symbol"].unique()
    for i in range(len(intra_companies)):
        for j in range(i, len(intra_companies)):
            inner_edge.append((name2index[intra_companies[i]], name2index[intra_companies[j]]))
```

```python
inner_edge
```

```python
len(inner_edge)
```

```python
inner_edge = np.array(inner_edge)
print(inner_edge.shape)
```

```python
np.save("./datasets4/inner_edge.npy", inner_edge)
```
