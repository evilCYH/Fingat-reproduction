```python
import torch
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 假设 edges 是你的 PyTorch 张量
inner_edge = np.array(np.load("./datasets/inner_edge.npy"))
inner_edge
```

```python
# 将 PyTorch 张量转换为边的列表
edge_list = list(inner_edge)

# 创建一个无向图
G = nx.Graph()

# 向图中添加边
G.add_edges_from(edge_list)

# 绘制图
plt.figure(figsize=(12, 12))
nx.draw(G, node_size=20, edge_color="gray")
plt.show()
```
